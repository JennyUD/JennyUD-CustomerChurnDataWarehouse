```python
import os
import zipfile

# Navigate to raw data folder
raw_path = r"C:\Users\igboa\Desktop\customer-churn-data-warehouse\data\raw"



zip_file = "customer-churn-dataset.zip"

with zipfile.ZipFile(zip_file, 'r') as z:
    z.extractall(raw_path)

print("Extracted files:", os.listdir(raw_path))
```

```python
import pandas as pd
from IPython.display import display, Markdown

train_path  = r"C:\Users\igboa\Desktop\customer-churn-data-warehouse\data\raw\customer_churn_dataset-training-master.csv"
test_path = r"C:\Users\igboa\Desktop\customer-churn-data-warehouse\data\raw\customer_churn_dataset-testing-master.csv"

train_df = pd.read_csv(train_path)
test_df  = pd.read_csv(test_path)


display(Markdown("### TRAIN DATASET"))
display(train_df.head())

display(Markdown("### TEST DATASET"))
display(test_df.head())


```

```python
train_df.info()
train_df.describe(include='all')
train_df.isna().sum()
train_df.nunique()
```

```python
full_df = pd.concat([train_df, test_df], ignore_index=True)
full_df.head()
```

```python
full_df.isna().sum()
```

```python
full_df[full_df.isna().any(axis=1)]
```

```python
clean_df = full_df.dropna()
```

```python
clean_df.isna().sum()
```

```python
duplicate_rows = clean_df[clean_df["CustomerID"].duplicated(keep=False)]
duplicate_rows.sort_values(by="CustomerID", inplace=True)
duplicate_rows.head(10)

```

```python
exact_dupes = (
    clean_df[clean_df.duplicated(keep=False)]
    .sort_values(list(clean_df.columns))
)

exact_dupes
```

```python
clean_df.to_csv("../../data/processed/clean_churn.csv", index=False)
```
